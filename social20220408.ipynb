```python
import requests
from bs4 import BeautifulSoup

# 특정 URL에 접속하는 요청(Request) 객체를 생성.
request = requests.get("http://www.cu.ac.kr/plaza/notice/notice")

# 접속한 이후의 웹 사이트 소스코드를 추출.
html = request.text

# HTML 소스코드를 파이썬 객체로 변환.
soup = BeautifulSoup(html, 'html.parser')

# <tr> 태그를 포함하는 요소를 추출
trs = soup.select('div.board_list tr')
print(len(trs))

# 모든 tr에 하나씩 접근.
board_data = []

for tr in trs:
    tds = tr.find_all('td')

    # 테이블 헤더는 스킵합니다.
    if len(tds) == 0:
        continue

    board_data.append(
        [tds[1].text.strip(), 
        tds[3].text.strip(), 
        tds[4].text.strip(), 
        tds[5].text.strip()])

import pandas as pd
board_df = pd.DataFrame(board_data, columns = ['title', 'writer', 'date', 'viewers'])
board_df
```
